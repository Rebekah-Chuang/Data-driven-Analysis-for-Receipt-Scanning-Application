```python
import pandas as pd
import numpy as np
```

### Show the first 5 rows for each table

```python
brands = pd.read_csv('brands.csv')
brands.head()
```

```python
receipts = pd.read_csv('receipts.csv')
receipts.head()
```

```python
receipt_items = pd.read_csv('receipt_items.csv')
receipt_items.head()
```

```python
users = pd.read_csv('users.csv')
users.head()
```

### Clean table brands

```python
#drop columns that are not used in brands table
brands.drop(columns = ['ID',
                       'CPG_ID',
                       'ROMANCE_TEXT',
                       'RELATED_BRAND_IDS'],
                        axis = 1, inplace = True)
brands.head()
```

```python
brands.set_index('BARCODE', inplace = True)
brands.head()
```

### Clean table receipts

```python
#drop columns that are not used in receipts table
receipts.drop(columns = ['USER_VIEWED',
                         'PENDING_DATE',
                         'MODIFY_DATE',
                         'FLAGGED_DATE',
                         'PROCESSED_DATE',
                         'NEEDS_FETCH_REVIEW',
                         'DELETED',
                         'NON_POINT_EARNING_RECEIPT'],
                          axis = 1, inplace = True)
receipts.head()
```

```python
receipts.set_index('ID', inplace = True)
receipts.head()
```

```python
#change the data type for PURCHASED_ITEM_COUNT into int
receipts['PURCHASED_ITEM_COUNT'] = receipts['PURCHASED_ITEM_COUNT'].astype(float).astype('Int64')
receipts.head()
```

### Clean table receipt_items

```python
receipt_items.drop(columns = ['REWARDS_RECEIPT_ITEM_ID',
                              'POINTS_EARNED',
                              'REWARDS_GROUP',
                              'ORIGINAL_RECEIPT_ITEM_TEXT',
                              'MODIFY_DATE'],
                               axis = 1, inplace = True)
receipt_items.head()
```

```python
#change the data type for BARCODE into int
receipt_items['BARCODE'] = receipt_items['BARCODE'].astype(float).astype('Int64')
receipt_items.dtypes
```

```python
#change the data type for QUANTITY_PURCHASED into int
receipt_items['QUANTITY_PURCHASED'] = receipt_items['QUANTITY_PURCHASED'].astype(float).astype('Int64')
receipt_items.head()
```

```python
receipt_items.set_index('REWARDS_RECEIPT_ID', inplace = True)
receipt_items.head()
```

### Clean table users

```python
users.drop(columns = ['LAST_REWARDS_LOGIN'],
                       axis = 1, inplace = True)
users.head()
```

```python
users.set_index('ID', inplace = True)
users.head()
```

### Save new csv file

```python
brands.to_csv('brands_2.csv')
receipts.to_csv('receipts_2.csv')
receipt_items.to_csv('receipt_items_2.csv')
users.to_csv('users_2.csv')
```
